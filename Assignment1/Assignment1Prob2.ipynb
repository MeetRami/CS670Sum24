```python
# Problem 2A
#Imports
import numpy as np
import matplotlib.pyplot as plt

#Given matrix
matrix = np.array([[4, 2, 1],
                   [2, 3, 1.5],
                   [1, 1.5, 2]])

# Sample size
m = 1000
# Mean is 0 for each dimension
mean = np.zeros(3)
# Transpose to get each vector as a column
samples = np.random.multivariate_normal(mean, matrix, m).T

# Performing Singular Value Decomposition (SVD) on the covariance matrix
U, s, Vt = np.linalg.svd(matrix)

# Taking the first two left singular vectors as principal components
firstTwoComponents = U[:, :2]
print(firstTwoComponents)

# Project the simulated vectors onto the subspace spanned by the first two principal components
projectedVec = np.dot(firstTwoComponents.T, samples)
print(projectedVec)
```

```python
# Problem 2B

#Convert U to normal array
U = np.array(U)

# Plot the vectors
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.quiver(0, 0, 0, firstTwoComponents[0][0], firstTwoComponents[1][0], firstTwoComponents[2][0], color='r', label='F P Component')
ax.quiver(0, 0, 0, firstTwoComponents[0][1], firstTwoComponents[1][1], firstTwoComponents[2][1], color='b', label='S P Component')

# Set plot limits
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set legend
ax.legend()

plt.title('Projected Vectors')
plt.show()
```

```python
#Problem 2C
import numpy as np
import matplotlib.pyplot as plt

# Reverse the projection to map back to the original 3D space
origVector = np.dot(firstTwoComponents, projectedVec)

# Plot the reconstructed points
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(origVector[0], origVector[1], origVector[2], color='b', alpha=0.5)

# Set plot limits
ax.set_xlim([-7, 7])
ax.set_ylim([-7, 7])
ax.set_zlim([-7, 7])

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Points Reconstructed in Original 3D Space')
plt.show()
```
